<h4>Utility Functions</h4>

```python
def plot(data, title, labels):
    
    plt.figure(figsize=(16,6)) 
    plt.title(title)
    
    for i in range(0, len(labels)):
        label = labels[i]
        (data[label] * (i+1)).plot(linewidth=2)

    plt.legend()
```
